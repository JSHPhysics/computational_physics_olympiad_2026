## 2. For loops

A "for loop" is a way to repeat the same code many times. This is one of the most useful tools in programming — it lets you do thousands of calculations without writing out thousands of lines of code.

The line `for n in range(1, 6):` means "give me each whole number from 1 up to (but **not** including) 6, one at a time, and call it `n`". So during the loop, `n` becomes 1, then 2, then 3, then 4, then 5.

Notice the indentation (the spaces at the start of the indented lines) — Python uses indentation to know which lines are "inside" the loop and should be repeated.

```python
# This loop adds up all the numbers from 1 to 5
total = 0                  # start at zero
for n in range(1, 6):      # n will be 1, 2, 3, 4, 5
    total = total + n      # add n to total each time round
print("Sum of 1 to 5 =", total)
# When you click play, the output should be 15  (because 1+2+3+4+5 = 15)


# Now you try:
# Can you modify the code below to count DOWN from 10 to 1, printing each number?
# Hint: range(start, stop, step) — try giving it a step of -1 to count backwards.
# Hint 2: don't forget the stop value is one PAST where you want to finish.
for i in range(1, 11):     # change this line!
    print(i)
```

## 4. NumPy arrays — lists of numbers that do maths

A normal Python list of numbers like `[1, 2, 3]` doesn't really know how to do maths. If you tried `[1, 2, 3] * 2`, Python would give you `[1, 2, 3, 1, 2, 3]` — it just repeats the list!

A NumPy array looks similar but every maths operation is applied to every number at once. This is called **vectorisation**, and it's one of the main reasons NumPy is so popular for science. It also makes the code look more like the maths you'd write on paper.

```python
# Create a NumPy array from a list of numbers
x = np.array([1, 2, 3, 4, 5])

# Watch what happens when we do maths with the array:
print("x       =", x)         # [1 2 3 4 5]
print("x * 2   =", x * 2)     # [ 2  4  6  8 10]   every element doubled
print("x ** 2  =", x ** 2)    # [ 1  4  9 16 25]   every element squared
print("sqrt(x) =", np.sqrt(x))  # square root of each element


# Now you try:
# Create your own array containing the numbers 10, 20, 30, 40, 50.
# Then divide every element by 10 and print the result.
my_array = np.array([])    # add the numbers between the square brackets
# print(my_array / 10)     # uncomment this line once you've filled in the array
```

## 5. Creating ranges of values

When plotting a smooth curve, you need lots of x-values close together. Typing them by hand would take forever, so NumPy gives us two tools:

- **`np.linspace(start, stop, N)`** — gives `N` evenly-spaced numbers from `start` to `stop` (including both ends).
- **`np.arange(start, stop, step)`** — gives numbers starting at `start`, increasing by `step` each time, stopping just before `stop`.

Use `linspace` when you know how many points you want (e.g. "100 points for a smooth curve"). Use `arange` when you know the step size (e.g. "every 0.1 from 0 to 10").

```python
# 5 evenly-spaced points between 0 and 10 (inclusive)
x = np.linspace(0, 10, 5)
print("linspace:", x)
# Output: [ 0.   2.5  5.   7.5 10. ]

# Whole numbers from 1 up to (not including) 6
n = np.arange(1, 6)
print("arange:  ", n)
# Output: [1 2 3 4 5]


# Now you try:
# Create an array of 11 evenly-spaced points between -5 and 5 using linspace.
# (You should get [-5, -4, -3, ..., 4, 5].)
my_x = np.linspace(0, 0, 0)    # change the three numbers!
print(my_x)
```

## 11. Vertical and horizontal reference lines

Sometimes you want to draw a horizontal or vertical line on your graph to highlight a special value — like the cut-off frequency for the photoelectric effect, or to show where y=0. Two functions do this:

- **`plt.axhline(y_value)`** — draws a horizontal line at that y-value
- **`plt.axvline(x_value)`** — draws a vertical line at that x-value

Both can take a `color`, `linestyle` (e.g. `"--"` for dashed), and `label` argument.

```python
# Photoelectric effect for copper (work function W = 4.7 eV)
f = np.linspace(0, 2e15, 200)    # frequencies from 0 to 2×10¹⁵ Hz
W_eV = 4.7
V = (constants.h / constants.e) * f - W_eV   # stopping voltage

plt.figure()
plt.plot(f, V, label="Stopping voltage")

# A horizontal line at V = 0 (where the curve crosses the x-axis)
plt.axhline(0, color="black", linewidth=0.5)

# A vertical dashed line at the cut-off frequency
cutoff = W_eV * constants.e / constants.h
plt.axvline(cutoff, color="purple", linestyle="--", label="cut-off frequency")

plt.xlabel("Frequency /Hz")
plt.ylabel("Stopping voltage /V")
plt.legend()
plt.grid(True)
plt.show()
# The cut-off frequency is where the line crosses zero — that's the minimum
# frequency of light needed to knock electrons out of copper.
```

## 12. Subplots — multiple plots side by side

When you want two graphs next to each other (or in a grid), use `plt.subplots`. The syntax looks unusual at first because we have to give names to both the whole figure **and** each individual sub-plot.

The line `fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))` means:

- "Make a figure with **1 row and 2 columns** of plots."
- "Call the whole figure `fig`, the left plot `ax1`, and the right plot `ax2`."
- "Make the figure 10 inches wide and 4 inches tall."

To draw on a specific subplot, use `ax1.plot(...)` instead of `plt.plot(...)`. The label commands also change slightly — they become `ax1.set_xlabel(...)` rather than `plt.xlabel(...)`. It's a bit annoying, but it's how matplotlib works.

```python
x = np.linspace(0, 10, 200)

# Make one figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Draw on the LEFT subplot (ax1)
ax1.plot(x, np.exp(-x))
ax1.set_title("Exponential decay")
ax1.set_xlabel("x")
ax1.grid(True)

# Draw on the RIGHT subplot (ax2)
ax2.plot(x, np.sin(x))
ax2.set_title("Sine wave")
ax2.set_xlabel("x")
ax2.grid(True)

plt.tight_layout()    # stops the labels from overlapping
plt.show()


# Now you try:
# Change the second subplot to show np.cos(x) instead of np.sin(x).
# Update the title to match.
```

## 13. 2D heatmaps — for orbital probability densities (Task 10)

For Task 10 you'll need to draw the probability density of an electron in 2D — a coloured map where bright colours mean "high chance of finding the electron here". This needs three things:

1. A **2D grid** of (x, y) coordinates covering the area you want to plot.
2. A **value** calculated at every point on the grid.
3. A way to **draw** the result as a coloured image.

`np.meshgrid` builds the grid for us, and `plt.imshow` draws it. Don't worry if `meshgrid` feels confusing — for now, just copy the pattern below and change the formula for `Z`.

```python
# Step 1: 1D arrays of x and y values
x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)

# Step 2: turn them into a 2D grid
# X and Y will both be 300x300 arrays — every grid point gets an x and a y
X, Y = np.meshgrid(x, y)

# Step 3: calculate something at every grid point
# This example is a "Gaussian bump" — peaked in the middle, fading outwards
Z = np.exp(-(X**2 + Y**2))

# Step 4: draw it as a coloured image
plt.figure(figsize=(6, 5))
plt.imshow(Z, extent=[-3, 3, -3, 3], origin="lower", cmap="viridis")
# extent tells imshow what the actual x,y limits are
# origin="lower" puts (0,0) at the bottom-left corner where you'd expect it
# cmap is the colour scheme — "viridis" is the default; try "plasma", "hot", "inferno"
plt.colorbar(label="probability density")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Example 2D probability density")
plt.show()


# Now you try:
# Change cmap to "plasma" and re-run the cell. Then try "hot" and "inferno".
# Which one looks most like an atomic orbital plot to you?
```

## 14. Building results inside a loop

A very common pattern: you have a quantity you want to calculate at many input values, and you want to plot the results. There are two ways to do this — looping (works for anything) and vectorising (faster but only works for simple maths). Here's both, doing the same calculation.

### The loop approach

We start with an empty list, then go through each input value, calculate the result, and add it to the list.

```python
# Calculate the de Broglie wavelength of electrons at different voltages
voltages = np.linspace(1000, 5000, 50)    # 1 kV to 5 kV, 50 points
wavelengths = []                          # start with an empty list

for V in voltages:
    p = np.sqrt(2 * constants.m_e * constants.e * V)   # momentum
    wavelengths.append(constants.h / p)                # add to the list

wavelengths = np.array(wavelengths)    # convert list to array for plotting

plt.figure()
plt.plot(voltages, wavelengths * 1e12)    # convert metres to picometres
plt.xlabel("Accelerating voltage /V")
plt.ylabel("de Broglie wavelength /pm")
plt.title("Loop version")
plt.grid(True)
plt.show()
```

### The vectorised approach (much shorter!)

For maths that NumPy already knows how to do, you can skip the loop entirely. The whole calculation runs on the array in one go.

```python
# Same calculation, but no loop — NumPy handles every voltage at once
p = np.sqrt(2 * constants.m_e * constants.e * voltages)
wavelengths = constants.h / p

plt.figure()
plt.plot(voltages, wavelengths * 1e12)
plt.xlabel("Accelerating voltage /V")
plt.ylabel("de Broglie wavelength /pm")
plt.title("Vectorised version (same result!)")
plt.grid(True)
plt.show()

# Use whichever style you find clearer. For Task 2 (Brownian motion), where each
# step depends on the previous one, you'll NEED the loop version — there's no
# way to vectorise a step-by-step simulation.
```

## 15. Animations in Colab

Colab can't open separate windows on your screen, so animations get embedded as videos right inside the notebook. The recipe has three parts:

1. Build a figure and an "artist" — something that will move (here, a single point).
2. Write an `update(frame)` function that says where the artist should be on each frame.
3. Use `FuncAnimation` to assemble all the frames, then `HTML(ani.to_jshtml())` to display it.

This is the most advanced section in the notebook — for now, treat the example below as a template you can copy and adapt for Task 2 (Brownian motion).

```python
from matplotlib import animation
from IPython.display import HTML

# Build the figure and the moving point
fig, ax = plt.subplots(figsize=(4, 4))
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect("equal")    # so circles look like circles, not ovals
ax.grid(True)
point, = ax.plot([], [], "ro", markersize=12)    # red circle, starts empty

# This function tells matplotlib where the point should be on each frame
def update(frame):
    angle = 2 * np.pi * frame / 60     # 60 frames = one full circle
    point.set_data([np.cos(angle)], [np.sin(angle)])
    return point,                       # the comma at the end is important!

# Assemble the animation
ani = animation.FuncAnimation(fig, update, frames=60, interval=50, blit=True)
plt.close()    # IMPORTANT: stops a static frame from also appearing above

# Display the animation as an embedded video with playback controls
HTML(ani.to_jshtml())
```

## 16. Interactive sliders with ipywidgets

For Challenge 8 (the visual calculator) you'll want sliders that let you change values and see the plot update live. The `ipywidgets` library makes this easy with the `interact` function — it automatically connects sliders to a function that draws something.

```python
from ipywidgets import interact, FloatSlider

# A function that takes some inputs and draws a plot
def plot_wave(amplitude, frequency):
    x = np.linspace(0, 2*np.pi, 200)
    y = amplitude * np.sin(frequency * x)
    plt.figure(figsize=(6, 3))
    plt.plot(x, y)
    plt.ylim(-3, 3)        # fix the y-axis so it doesn't jump around
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.show()

# Connect the function to two sliders
interact(plot_wave,
         amplitude=FloatSlider(min=0, max=3, step=0.1, value=1),
         frequency=FloatSlider(min=0.5, max=5, step=0.1, value=1));
# When you run this cell, drag the sliders to see the plot update live!


# Now you try:
# Add a third slider called "phase" that shifts the wave left/right.
# You'll need to:
#   1. Add 'phase' as an argument to plot_wave(...)
#   2. Use it in the formula: y = amplitude * np.sin(frequency * x + phase)
#   3. Add a phase=FloatSlider(...) to the interact call (try min=0, max=2*np.pi)
```

## 17. Special functions for hydrogenic orbitals (Task 10)

⚠️ **Argument order matters** for `sph_harm_y`: it takes `(l, m, theta, phi)` where `theta` is the polar angle (between 0 and π) and `phi` is the azimuth (between 0 and 2π). If you find older code online using `sph_harm(...)`, that's a deprecated version — use `sph_harm_y` instead.

```python
# Associated Legendre polynomial P_l^m(x)
print("P_2^0(0.5)    =", special.lpmv(0, 2, 0.5))

# Spherical harmonic Y_l^m(theta, phi)
# Argument order: (l, m, theta=polar, phi=azimuth)
print("Y_1^0(pi/2, 0)=", special.sph_harm_y(1, 0, np.pi/2, 0.0))

# Generalised (associated) Laguerre polynomial L_n^k(x)
print("L_2^1(1)      =", special.assoc_laguerre(1.0, 2, 1))

# Factorial — useful in many quantum mechanics formulas
print("5!            =", special.factorial(5))
```

## 19. Common pitfalls (read this when something goes wrong!)

**`turtle` doesn't work in Colab.** The `turtle` library needs a desktop window, but Colab runs on a remote server with no screen attached. You'll get the error `TclError: no display name and no $DISPLAY environment variable`. Use `matplotlib` for any drawing instead.

**Counting starts at 0.** Python and NumPy count from zero, not one. So `x[0]` is the *first* element of `x`, and `range(5)` gives `0, 1, 2, 3, 4` — five values starting from zero, not 1 to 5. This trips up almost everyone at first.

**Indentation matters.** Lines that go "inside" a loop, function, or `if`-statement must be indented (use 4 spaces, or one tab — be consistent). If you get an `IndentationError`, check that all the indented lines line up the same way.

**Restart Runtime if things go strange.** A cell that errored partway through can leave variables in a confusing state. Use **Runtime → Restart runtime** to start fresh. Don't worry — your code is still saved, only the running state is reset.

**Idle disconnect.** Free Colab disconnects you after about 90 minutes of inactivity. Save to Drive often (**File → Save a copy in Drive**).

**Animation static frame.** Always call `plt.close()` between building a `FuncAnimation` and calling `HTML(ani.to_jshtml())`, otherwise the first frame of the animation appears as a separate static plot above the video.

**Read the error message.** When something breaks, Python tells you exactly what went wrong and which line caused it. The error usually starts with the type (e.g. `NameError`, `TypeError`, `IndexError`) and includes a line number — read these carefully before asking for help.
